# AI Model Performance Analysis
This notebook demonstrates how to:
- Find and load JSON logs generated by `benchmark.py`
- Plot inference times, CPU usage, and other metrics using `matplotlib`

## Prerequisites
Make sure you have generated logs (e.g., `metrics_log_resnet18_YYYYMMDD_HHMMSS.json`) in a `logs/` folder by running:
```bash
python benchmark.py --model resnet18 --iterations 20 --save-logs
```

Then place this notebook (`analysis.ipynb`) in the same directory as your `logs/` folder.

````python
import os
import json
import matplotlib.pyplot as plt
  # For inline plotting in Jupyter
````

## 1. Helper Functions

````python
def find_log_files(log_dir='logs'):
    """
    Returns a list of JSON log files in the specified directory.
    """
    if not os.path.exists(log_dir):
        return []

    files = [f for f in os.listdir(log_dir)
             if f.startswith('metrics_log_') and f.endswith('.json')]
    return sorted(files)

def load_logs(filepath):
    """Load a single JSON file containing benchmark logs."""
    with open(filepath, 'r') as f:
        data = json.load(f)
    return data
````

## 2. List Available Log Files

````python
log_files = find_log_files()
print("Found the following log files:")
for lf in log_files:
    print(lf)
````

## 3. Choose a Log File to Analyze
Manually set the `selected_log` variable to one of the files above.

````python
# Example: If you see 'metrics_log_resnet18_20250123_120000.json' in the list above,
# set selected_log = 'metrics_log_resnet18_20250123_120000.json'

selected_log = 'metrics_log_alexnet_20250126_071410.json'  # Replace with a filename from the list above

if selected_log is None:
    print("Please set 'selected_log' to a valid filename from the 'logs/' folder.")
````

## 4. Plot Inference Times and CPU Usage
This code block loads the chosen file and plots:
- Inference time vs. iteration
- CPU usage before vs. after inference

````python
if selected_log and os.path.isfile(os.path.join('logs', selected_log)):
    # Load logs
    file_path = os.path.join('logs', selected_log)
    logs_data = load_logs(file_path)
    print(f"Loaded {len(logs_data)} entries from {selected_log}")

    # Extract iteration, inference time, CPU usage
    iterations = [entry['iteration'] for entry in logs_data]
    inference_times = [entry['inference_time'] for entry in logs_data]
    cpu_before = [entry['before']['cpu_percent'] for entry in logs_data]
    cpu_after = [entry['after']['cpu_percent'] for entry in logs_data]
    mem_before = [entry['before']['memory_percent'] for entry in logs_data]
    mem_after = [entry['after']['memory_percent'] for entry in logs_data]

    # 4.1 Plot Inference Times
    plt.figure(figsize=(10, 5))
    plt.plot(iterations, inference_times, marker='o')
    plt.title('Inference Time per Iteration')
    plt.xlabel('Iteration')
    plt.ylabel('Time (seconds)')
    plt.grid(True)
    plt.show()

    # 4.2 Plot CPU Usage Before/After
    plt.figure(figsize=(10, 5))
    plt.plot(iterations, cpu_before, 'r--o', label='CPU Before')
    plt.plot(iterations, cpu_after, 'b--o', label='CPU After')
    plt.title('CPU Usage Before & After Inference')
    plt.xlabel('Iteration')
    plt.ylabel('CPU Usage (%)')
    plt.legend()
    plt.grid(True)
    plt.show()

    # 4.3 (Optional) Plot Memory Usage Before/After
    plt.figure(figsize=(10, 5))
    plt.plot(iterations, mem_before, 'g--o', label='Memory Before')
    plt.plot(iterations, mem_after, 'm--o', label='Memory After')
    plt.title('Memory Usage Before & After Inference')
    plt.xlabel('Iteration')
    plt.ylabel('Memory Usage (%)')
    plt.legend()
    plt.grid(True)
    plt.show()

else:
    print("No valid log file selected or file does not exist.")
````
